## Changing Spatial Resolution of Image
For every 3×3 block of the image (without overlapping), replace all corresponding 9 pixels by their average. This operation simulates reducing the image spatial resolution. Repeat this for 5×5 blocks and 7×7 blocks.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from math import floor
```

```python
img = cv2.imread('images/hermmine.jpg',0)
img.shape
plt.imshow(img, cmap='gray')
plt.show()

# chunk = img[:300,0:300]
# plt.imshow(chunk, cmap='gray')
# plt.show()
```

```python
# Resizing the image to match with the N parameter
n = 5
img_re = cv2.resize(img, (floor(img_re.shape[1]/n)*n,floor(img_re.shape[0]/n)*n))
plt.imshow(img_re, cmap='gray')
plt.show()
```

```python
img_av = np.zeros(img_re.shape)

for row in range(floor(img_re.shape[0]/n)):
    for col in range(floor(img_re.shape[1]/n)):
        img_av[row*n:(row+1)*n, col*n:(col+1)*n].fill(np.mean(img_re[row*n+1:(row+1)*n, col*n+1:(col+1)*n]))
        
plt.imshow(img_av, cmap='gray')
plt.show()
```

## Rotation of Image 


```python
img = cv2.imread('images/pencil.jpg')

rows, cols = img.shape[0:2] # Its a three channel image. So take first two coloumns 

M = cv2.getRotationMatrix2D((cols/2, rows/2), 45, 1.5) # 35 degree rotation with 1.5x scaling
img_rotated = cv2.warpAffine(img, M, (cols, rows))
```

```python
plt.imshow(img, cmap='gray')
plt.show()
plt.imshow(img_rotated, cmap='gray')
plt.show()
```

## Averaging Image 

load an image and then perform a simple spatial 3x3 average of image pixels. In other words, replace the value of every pixel by the average of the values in its 3x3 neighborhood.

```python
img = cv2.imread('images/keira.jpg',0)

rows, cols = img.shape

plt.imshow(img, cmap='gray')
plt.show()
```

```python
img2 = np.zeros(img.shape)
for row in range(rows):
    for col in range(cols):
        if row-1<0 and col-1<0:
            img2[row,col] = np.mean(img[row:row+1, col: col+1])
        elif row-1<0:
            img2[row,col] = np.mean(img[row:row+1, col-1: col+1])
        elif col-1<0:
            img2[row,col] = np.mean(img[row-1:row+1, col: col+1])
        else:
            img2[row,col] = np.mean(img[row-1:row+1, col-1: col+1])
        
```

```python
plt.imshow(img2, cmap='gray')
plt.show()
```

## Change Image Intesity Levels 
Write a computer program capable of reducing the number of intensity levels in an image from 256 to 2, in integer powers of 2.

```python
n = 64
img3 = np.zeros(img.shape)
for row in range(rows):
    for col in range(cols):
        img3[row,col] = floor(img[row,col]/n)*n

plt.imshow(img, cmap='gray')
plt.show()
plt.imshow(img3, cmap='gray')
plt.show()

# Unique values in img3
print(np.unique(img3))
# Unique values in img
print(np.unique(img).size)
```

```python
img4 = cv2.resize(img,(floor(img.shape[1]/8)*8,floor(img.shape[0]/8)*8))

total_blocks = (img4.shape[0]*img4.shape[1])/64

plt.imshow(img4, cmap='gray')
plt.show()

print(total_blocks)
print(img4.shape)
```

```python
rows, cols = img4.shape
i = 0
block = np.zeros((int(total_blocks), 8,8))
for row in range(int(rows/8)):
    for col in range(int(cols/8)):
        block[i,:,:] = img4[row*8:(row+1)*8, col*8:(col+1)*8]
        i=i+1
```

```python
block_dct = np.zeros(block.shape)
i = 0
for b in block:
    block_dct[i,:,:] = cv2.dct(b)
    i=i+1
```

```python
n = 8
block_quant = (np.floor_divide(block_dct,n))*n
```

```python
block_idct = np.zeros(block.shape)
i = 0
for b in block_quant:
    block_idct[i,:,:] = cv2.idct(b)
    i=i+1
```
